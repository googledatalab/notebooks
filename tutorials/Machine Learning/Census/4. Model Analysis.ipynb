## Model Analysis

Analyze the model by looking at the summary events the training program outputs. Let's look at the events the local run generated in previous notebook, but events generated by cloud run in GCS works too.<br>

First list events under certain directories.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/census/model/summaries /content/datalab/tmp/ml/census/model/eval
```

Plot events.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/census/model/summaries  /content/datalab/tmp/ml/census/model/eval  --name loss error --step
```

Note that you can plot events in the same graph from multiple directories (--dir support multiple values, just like --name. --dir also supports "*" in the path and it will exapnd the dir name). This is useful if you want to compare two model's training progress. Below we plot events under two different model dirs: /content/datalab/tmp/ml/census/model/ and /content/datalab/tmp/ml/census/largermodel/.<br>

```python
%mlalpha summary --dir /content/datalab/tmp/ml/census/*/summaries /content/datalab/tmp/ml/census/*/eval --name loss error --step
```
