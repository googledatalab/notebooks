## Model Analysis

Analyze the model by looking at the summary events the training program outputs. Let's look at the events the two local run generates in previous notebook, but events generated by cloud run works too.

Using "%mlalpha summary" command to view events. First list events in a given dir.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/iris/model/train
```

Plot loss graphs one based on time and one based on time.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/iris/model/train --name loss --time
```

Plot loss graph with steps, and also compare two runs.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/iris/model/train  /content/datalab/tmp/ml/iris/model_lr/train --name loss --step
```

Multiple events can be plotted in the same graph too.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/iris/model/train  /content/datalab/tmp/ml/iris/model_lr/train --name loss dnn/hiddenlayer_0:fraction_of_zero_values --step
```

Accuracy is boring (only one point per run) since we do it only once at the end of the training.

```python
%mlalpha summary --dir /content/datalab/tmp/ml/iris/model*/train/eval  --name accuracy --step
```

Note that the summary dir can be GCS path too.
